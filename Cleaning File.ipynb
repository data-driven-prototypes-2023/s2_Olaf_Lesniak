```python
import pandas as pd
import numpy as np

# load data
df = pd.read_csv('https://raw.githubusercontent.com/data-driven-prototypes-2023/s2_Olaf_Lesniak/fe1f65d2805c384683dcb3ed3d46e59dc5cbd357/openfoodfacts_export_us.csv', sep='\t')
```

```python
# We have discovered that there are 209 columns but most of them are empty with zero or NaN values --> drop them
df = df.dropna(axis=1, how='all') #drop all columns that that only contain zero or NAN values
df.describe()
```

```python
# First preselection of columns based on excel exploration
selected_columns = [
    'code', 'product_name_en', 'brands', 'brands_tags', 'quantity', 'serving_size', 
    'categories', 'categories_tags', 'labels', 'labels_tags', 'countries_tags', 
    'origins', 'origins_tags', 'manufacturing_places', 'emb_codes', 'allergens', 
    'allergens_tags', 'traces', 'traces_tags', 'energy-kcal_unit', 'energy-kcal_value', 
    'no_nutrition_data', 'nutrition_data_per', 'fat_value', 'fat_unit', 'saturated-fat_value', 
    'saturated-fat_unit', 'carbohydrates_value', 'sugars_value', 'carbohydrates_unit', 
    'proteins_value', 'proteins_unit', 'salt_value', 'salt_unit', 'sodium_value', 
    'sodium_unit', 'energy_value', 'energy_unit', 'off:food_groups', 'off:food_groups_tags', 
    'off:nutriscore_grade', 'off:nutriscore_score', 'off:ecoscore_score', 'off:ecoscore_grade'
]

df = df[selected_columns]
df.count()
```

```python
import pandas as pd
from googletrans import Translator
# Data cleaning steps (had to separate do to the )


# Standardizing brand names (example: converting all to lower case and stripping whitespace)
df['brands'] = df['brands'].str.lower().str.strip()

# Cleaning up the values where we have grams listed and adjusting column headers
for col in df.columns:
    if ' (g)' in col:
        # Extracting the unit from the column name
        new_col_name = col.replace(' (g)', '')
        # Standardizing the values in the column to numeric, ignoring errors
        df[new_col_name] = pd.to_numeric(df[col], errors='coerce')
        # Dropping the old column
        df.drop(columns=[col], inplace=True)

# Dropping the prefix `en:` from all column names
df.columns = [col.replace('en:', '') for col in df.columns]

# Standardizing 'off:nutriscore_grade' values to lowercase
df['off:nutriscore_grade'] = df['off:nutriscore_grade'].str.lower()

# Dropping rows that don't have 'off:nutriscore_grade' as a, b, c, d, or e
valid_grades = ['a', 'b', 'c', 'd', 'e']
df = df[df['off:nutriscore_grade'].isin(valid_grades)]

# Translate categories column into English
translator = Translator()

def safe_translate(text, translator):
    try:
        if pd.notnull(text) and text.strip():
            return translator.translate(text, dest='en').text
        return text
    except Exception as e:
        return text

df['categories'] = df['categories'].apply(lambda x: safe_translate(x, translator))

# Split categories into a list
df['categories'] = df['categories'].str.split(',')

# Flatten the list of categories and get the top 6 most common categories
all_categories = df['categories'].explode()
top_categories = all_categories.value_counts().head(6).index.tolist()

# Create new columns for the top 6 categories
for category in top_categories:
    df[category] = df['categories'].apply(lambda x: 'Yes' if category in x else 'No')

# Drop the original 'categories' column
df.drop(columns=['categories'], inplace=True)

# Save the cleaned data to a new CSV file (so that we don't have to run the cleaning steps again and loading time was very long)
cleaned_file_path = 'https://raw.githubusercontent.com/data-driven-prototypes-2023/s2_Olaf_Lesniak/fe1f65d2805c384683dcb3ed3d46e59dc5cbd357/openfoodfacts_export_clean_selected_categories.csv'
df.to_csv(cleaned_file_path, sep=',', index=False)
```

```python
# Data Cleaning whole summary + converting values into standardized values [grams, ml, count]
import pandas as pd
import re

# display the whole df not skipping all columns
pd.set_option('display.max_columns', None)

# Dropping the prefix `en:` from all column names
df = df.applymap(lambda x: x.replace('en:', '') if isinstance(x, str) else x)

# Display unique values of a column and its frequency
print(df['brands'].value_counts())
print(df['brands'].unique())
print(df["energy_unit"].value_counts())

# Ensure serving_size column is treated as strings
df['serving_size'] = df['serving_size'].astype(str)

# Extract only the number of grams from the string marked as ()
df['serving_size'] = df['serving_size'].str.extract(r'\((\d+\.?\d*)\s*g\)', expand=False)
df['serving_size'] = pd.to_numeric(df['serving_size'], errors='coerce')

# Conversion factors
oz_to_g = 28.3495
lb_to_g = 453.592
kg_to_g = 1000
fl_oz_to_ml = 29.5735
quart_to_ml = 946.353
pint_to_ml = 473.176

# Function to convert quantity to grams or milliliters
def convert_quantity(quantity):
    try:
        quantity = quantity.strip().lower()
        if 'g' in quantity:
            match = re.search(r'([\d\.]+)\s*g', quantity)
            if match:
                return float(match.group(1)), 'g'
        elif 'kg' in quantity:
            match = re.search(r'([\d\.]+)\s*kg', quantity)
            if match:
                return float(match.group(1)) * kg_to_g, 'g'
        elif 'oz' in quantity:
            match = re.search(r'([\d\.]+)\s*oz', quantity)
            if match:
                return float(match.group(1)) * oz_to_g, 'g'
        elif 'lb' in quantity:
            match = re.search(r'([\d\.]+)\s*lb', quantity)
            if match:
                return float(match.group(1)) * lb_to_g, 'g'
        elif 'l' in quantity:
            match = re.search(r'([\d\.]+)\s*l', quantity)
            if match:
                return float(match.group(1)) * 1000, 'ml'
        elif 'fl oz' in quantity:
            match = re.search(r'([\d\.]+)\s*fl\s*oz', quantity)
            if match:
                return float(match.group(1)) * fl_oz_to_ml, 'ml'
        elif 'ml' in quantity:
            match = re.search(r'([\d\.]+)\s*ml', quantity)
            if match:
                return float(match.group(1)), 'ml'
        elif 'quart' in quantity or 'qt' in quantity:
            match = re.search(r'([\d\.]+)\s*quart|qt', quantity)
            if match:
                return float(match.group(1)) * quart_to_ml, 'ml'
        else:
            return float(quantity), 'count'
    except ValueError:
        return None

# Ensure the quantity column is treated as strings
df['quantity'] = df['quantity'].astype(str)

# Convert all quantities, ensuring None values are handled
converted_quantities = df['quantity'].apply(convert_quantity)
df["quantity_converted"] = [x[0] if x is not None else None for x in converted_quantities]
df["unit"] = [x[1] if x is not None else None for x in converted_quantities]

# Conversion factors
mg_to_g = 0.001

# Function to convert salt value to grams
def convert_to_grams(value, unit):
    if unit == 'mg':
        return value * mg_to_g
    return value

# Apply the conversion
df['salt_value_in_grams'] = df.apply(lambda row: convert_to_grams(row['salt_value'], row['salt_unit']), axis=1)
df['sodium_value_in_grams'] = df.apply(lambda row: convert_to_grams(row['sodium_value'], row['sodium_unit']), axis=1)

# Select only based on specific values 
df = df[df['energy_unit'] == 'kcal']
df = df[df['nutrition_data_per'] == '100g']
df = df[df["unit"] == "count"]
df = df[df["countries_tags"] == "united-states"]
df = df[df['off:food_groups'] == 'milk-and-yogurt']

# Drop columns
df = df.drop(columns=['nutrition_data_per','quantity_converted','saturated-fat_unit','no_nutrition_data','traces_tags','labels_tags','brands_tags','countries_tags', 'energy_unit', 'off:food_groups','sodium_value', 'sodium_unit','salt_value', 'salt_unit','unit', 'quantity', 'origins', 'origins_tags', 'manufacturing_places', 'emb_codes', 'allergens', 'energy-kcal_unit', 'fat_unit', 'carbohydrates_unit', 'proteins_unit'])

# Renaming columns
df = df.rename(columns={
    'fat_value': 'fat_value_per_100g', 
    'saturated-fat_value': 'saturated_fat_value_per_100g', 
    'proteins_value': 'proteins_value_per_100g', 
    'carbohydrates_value': 'carbohydrates_value_per_g'
})

df = df.dropna(subset=['serving_size', 'saturated_fat_value_per_100g', 'off:ecoscore_score'])

df['traces'] = df['traces'].fillna('no traces')
df['allergens_tags'] = df['allergens_tags'].fillna('no_traces')
df['brands'] = df['brands'].fillna("Not Known Brand")
df['label'] = df['label'].fillna('No Label')

allergens_split = df["allergens_tags"].str.lower().str.get_dummies(sep=",")
df = pd.concat([df, allergens_split], axis=1)

# Round all numeric values to 2 decimal places
df = df.round(2)

# Display the result
df.head()
df.shape
```

```python
# Run a final IQR data cleaning throught the numerical values
def remove_outliers(df, factor=10): # more flexible to not kill too many outliers (based on experience)
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for column in numeric_cols:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

df =remove_outliers(df)

# Export cleaned DF as CSV which then will be used by the Dash-Jupyter Notebooks file for visualization
df.to_csv('C:/Users/Admin/Desktop/Esade/Prototype/Assignment2/openfoodfacts_yogurt_clean.csv', index=False)
df.shape
```
